```python
import json

with open('secret.json', 'r') as f:
    secrets = json.loads(f.read())
    
SECRET_KEY = secrets['Secret_key']['Google API']

name = []
rating = []
review = []
```

```python
import requests

for i in df['ID']:
    print(i)
    url = "https://maps.googleapis.com/maps/api/place/details/json?place_id="+str(i)+"&fields=name%2Crating%2Cformatted_address%2Creviews&language=ko&key="+str(SECRET_KEY)

    payload={}
    headers = {}

    response = requests.request("GET", url, headers=headers, data=payload)

    data = response.json()
    if 'reviews' in data['result']:
        for i in range(len(data['result']['reviews'])):
            name.append(data['result']['name'])
            rating.append(data['result']['reviews'][i]['rating'])
            review.append(data['result']['reviews'][i]['text'])
```

```python
import pandas as pd
data = pd.DataFrame({'name': name, 'rating': rating, 'review': review})
data
```
